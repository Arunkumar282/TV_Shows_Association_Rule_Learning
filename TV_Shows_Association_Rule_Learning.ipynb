# **IMPORTING LIBRARIES**

```python
# importing libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

# **IMPORTING THE DATASET**

```python
# Here we are loading netflix data file into the dataset using pandas.
dataSet = pd.read_csv("/TV Shows - Association Rule Learning.csv")
# Creating a list and appending the dataset in the form of string.
transactions = []
for i in range(0,len(dataSet)):
  transactions.append([str(dataSet.values[i,j]) for j in range(0,12)])
```

# **APRIORI TRAINING ON DATASET**

```python
# we are setting the values for the apriori which is a algorithm for associtaion rule mining from 'apyori' python library.
output = apriori(transactions = transactions,min_support=0.003,min_confidence=0.2,min_lift = 3,min_length=2,max_length=2)
# Now, we are converting into list and store it in the output variable and displaying the output.
output = list(output)
output
```

# **PROPER DISPLAY OF OUTPUT**

```python
# In the show funtion we are passing output in which it take in the basis of support values for the itemsets.
def show(ouput):
  column_1 = [tuple(singleTuple[2][0][0])[0] for singleTuple in output]
  column_2 = [tuple(singleTuple[2][0][1])[0] for singleTuple in output]
  support = [support[1] for support in output]
  return list(zip(column_1,column_2,support))
# Now, by using pandas we are creating/visualizing a table in which it shows the list of the frequent movies.
results = pd.DataFrame(show(output),columns = ['Movie_1','Movie_2','Support'])
# Simply, displaying the all frequent itemsets in the table.
results
```

```python
#Finally, we are displaying the top most watched frequent TV SHOWS from the dataset using APRIORI.
results.nlargest(n = 10,  columns = 'Support')
```
